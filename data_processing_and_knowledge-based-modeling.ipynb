```python
# Remove '$' and ',' from the MSRP(Min) column
import re

min = []

for i, row in remove_na_df.iterrows():
    x = int(re.sub("[^0-9]", "", row['MSRP(Min)']))
    print(x)
    min.append(x)

del remove_na_df['MSRP(Min)']
remove_na_df.insert(4, 'MSRP(Min)', min, allow_duplicates=True)
    
remove_na_df

    
```

```python
max = []
for i, row in remove_na_df.iterrows():
    x = re.sub("[^0-9]", "", row['MSRP(Max)'])
    max.append(x)

del remove_na_df['MSRP(Max)']
remove_na_df.insert(5, 'MSRP(Max)', max, allow_duplicates=True)
    
remove_na_df
```
