```python
from selenium import webdriver
import pandas as pd
import pprint

pp = pprint.PrettyPrinter(indent=4)

CAR_BRAND_NUM = 20
```

```python
#open website
driver = webdriver.Chrome()
base_url = "https://www.edmunds.com/"
driver.get(f"{base_url}acura/")

cars = {}

#Extract Car Manufacturers 
for i in range(1, CAR_BRAND_NUM+1):
    car_brands = driver.find_elements_by_xpath(f'/html/body/div[1]/div/main/div/div[2]/div[2]/div[1]/div/p[{i}]/a')[0]
    #print(car_brands.text)
    brand = car_brands.text
    cars[brand.lower()] = {}
    
print(cars)
driver.close()
```

```python
#Extract Car models listed on Edmunds under the Popular Brands and links and store it in python dictioanry

brand_models = {'acura': 8, 'audi': 32, 'bmw':46, 'buick':9, 'chevrolet':40, 
                'chrysler':3, 'dodge':15, 'fiat':9, 'ford':45, 'gmc':17, 'honda':20, 
                'hyundai':19, 'infiniti':11, 'jeep':9, 'kia':17, 'lexus':20, 'mazda':8,
                'subaru':11, 'toyota':26, 'volkswagen':15}
driver = webdriver.Chrome()

for make in cars:
    #print(f'{base_url}{make}/')
    driver.get(f'{base_url}{make}/')
        
    for i in range(1, brand_models[make]+1):
        try:
            models_element = driver.find_elements_by_xpath(f'/html/body/div[1]/div/main/div/div[2]/div[1]/ul[2]/li[{i}]/div/div[2]/h3/a')[0]
            cars[make][models_element.text] = {}
            
            #Extract hyperlink to each vehicle 
            cars[make][models_element.text]['link'] = models_element.get_attribute('href')
        except Exception as e:
            print(f"error for {make}:\n {e}")
   
    #Add year attribute to each vehicle 
    for i in cars[make]:
        x = i.split(' ')
        cars[make][i]['year'] = str(x[0])
        cars[make][i]['make'] = make

        #print(i)

pp.pprint(cars)
driver.close()
```

```python
#categorize each vehicle using term frequency and get MSRP value
import re
driver = webdriver.Chrome()

terms = ['suv', 'sedan', 'coupe', 'hatchback', 'truck', 'wagon', 'copnvertibles', 'minivan', 'electric', 'hybrid']

html_source = driver.page_source
_max = 0
category = ''
for make in _cars:
    for i in _cars[make]:
        driver.get(_cars[make][i]['link'])
        html_source=driver.page_source
        for _type in terms:
            x = re.findall(f"/({_type})+/", html_source)
            #print(f"{i} : {len(x)}")
            if(len(x) > _max):
                _max = len(x)
                category = _type
        _cars[make][i]['category'] = category

        _max = 0
        category = ''
        
        msrp = driver.find_elements_by_xpath('/html/body/div[1]/div/main/div[2]/div/div[1]/div[1]/div[1]/div[2]/div[1]/div')[0]
        msrp_t = msrp.text
        msrp_t = msrp_t.replace(' ', '')
        if '-' in msrp_t:
            msrp_t = msrp_t.split('-')
            msrp_min = msrp_t[0]
            msrp_max = msrp_t[1]
        else:
            msrp_min = msrp_t
            msrp_max = msrp_t

        _cars[make][i]['MSRP_Min'] = msrp_min
        _cars[make][i]['MSRP_Max'] = msrp_max

        msrp_min = None
        msrp_max = None
            
driver.close()
```
